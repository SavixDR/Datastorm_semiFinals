```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# extract the features
X = sales_grouped1[['BP1'    , 'BP2'     ,'BP3'     ,'BP4'     ,'BP5'    , 'IP1'     ,'IP2'     ,'IP3'  ,'IP4'     ,'IP5' , 'total_price' , 'Total_Volume_week' ,'Total_bulk_price_week'	,'Total_impulse_price_week' ,'Total_Bulk_Volume_week'  ,'Total_Impulse_Volume_week' ]]

# define the number of clusters
n_clusters = 5

# instantiate the KMeans model
kmeans = KMeans(n_clusters=n_clusters, random_state=42)

# fit the model to the data
kmeans.fit(X)

# predict the cluster labels for each outlet ID
sales_grouped1['cluster'] = kmeans.predict(X)

sales_grouped1.head(20)

```

```python
sales_grouped1['cluster'].unique()
```

```python
# create an empty dictionary to store the dataframes
cluster_dfs = {}

# iterate over each cluster value
for cluster in range(5):
    # filter the dataframe by the current cluster value
    cluster_df =sales_grouped1[sales_grouped1['cluster'] == cluster]
    
    # add the cluster dataframe to the dictionary
    cluster_dfs[f'cluster_{cluster}'] = cluster_df

# access the dataframe for cluster 0
cluster_0_df = cluster_dfs['cluster_0']
cluster_1_df = cluster_dfs['cluster_1']
cluster_2_df = cluster_dfs['cluster_2']
cluster_3_df = cluster_dfs['cluster_3']
cluster_4_df = cluster_dfs['cluster_4']

```

calculating the silhoutte coefficient

```python
from sklearn.metrics import silhouette_score

# select features for clustering
X = sales_grouped1[['BP1'    , 'BP2'     ,'BP3'     ,'BP4'     ,'BP5'    , 'IP1'     ,'IP2'     ,'IP3'  ,'IP4'     ,'IP5' , 'total_price' , 'Total_Volume_week' ,'Total_bulk_price_week'	,'Total_impulse_price_week' ,'Total_Bulk_Volume_week'  ,'Total_Impulse_Volume_week' ]]

# set number of clusters to evaluate
k_values = range(2, 10)

# calculate Silhouette Coefficient for each k value
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    print(f"Silhouette Coefficient for k={k}: {silhouette_avg:.3f}")
```

Calculating the Davis_bouldin Index

```python
from sklearn.metrics import davies_bouldin_score

# perform K-means clustering with k=4
kmeans = KMeans(n_clusters=5, random_state=42).fit(X)

# calculate Davies-Bouldin Index
dbi = davies_bouldin_score(X, kmeans.labels_)

print("Davies-Bouldin Index:", dbi)
```

Calculating the Calinski-Harabasz Index

```python
from sklearn.metrics import calinski_harabasz_score

# initialize k-means clustering with 3 clusters
kmeans = KMeans(n_clusters=5, random_state=42)

# fit k-means model on the data
kmeans.fit(X)

# calculate Calinski-Harabasz Index
ch_score = calinski_harabasz_score(X, kmeans.labels_)

print("Calinski-Harabasz Index:", ch_score)
```
